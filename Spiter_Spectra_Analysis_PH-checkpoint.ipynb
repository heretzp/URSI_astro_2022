```python
from spec_utils import read_irs
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
```

```python
# AS205 A
irs=read_irs('AS205A',noerror=True,dd=r'C:\Users\pheretz\Desktop\URSI_Internship\data\Spitzer\REDUCED\\')
fig=plt.figure(figsize=(15,10))
axl=fig.add_subplot(111)
axl.plot(irs.wave,irs.flux,label='flux')
axl.plot(irs.wave,irs.cont,label='cont')
axl.plot(irs.wave,irs.orig,label='orig')
axl.set_ylabel(r'F$_\nu$ [Jy]')
axl.set_xlabel('Wavelength [$\mu$m]')
axl.legend()
```
