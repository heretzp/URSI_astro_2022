```python
# Error bars practice

data = pd.read_csv (r'C:\Users\pheretz\Desktop\URSI Internship\Sources Info Sheet - Catalog.csv')
df = pd.DataFrame(data, columns= ['Star','disk_inclination','uncertainty_disk_inclination_positive','uncertainty_disk_inclination_negative'])
df_loc = df.loc[df['disk_inclination'] != -99.0]

print (df_loc)
```

```python
star_list = df_loc['Star'].tolist()

disk_inclination = df_loc['disk_inclination'].tolist()

y_error = [df_loc['uncertainty_disk_inclination_negative'].tolist()]
```

```python
fig=plt.figure(1,figsize=(18,10))
ax1=fig.add_subplot(111)
ax1.plot(star_list,disk_inclination, 'ro')
for i, (x, y) in enumerate(zip(star_list,disk_inclination)):
    ax1.text(x,y,star_list[i],fontsize=10)
plt.errorbar(star_list,disk_inclination, yerr = y_error,  
             ecolor=None, elinewidth=None, capsize=None, barsabove=False, lolims=False, uplims=False, 
             xlolims=False, xuplims=False, errorevery=1, capthick=None)
ax1.set_xlabel('Age (yrs)',fontsize=18)
ax1.set_ylabel('Number of Bright Rings',fontsize=18)

#ax1.set_xlim(1,4)   #Set limits for x axis to make the plot look better
#ax1.set_ylim(90,400)
```
